## Demo

```python
from hash_table import FixedHashTable
from book_dataset import BookDataset
```

```python
# Create dataset and get word frequencies
dataset = BookDataset()
word_freq = dataset.create()

# Take first 100 words
first_100 = dict(list(word_freq.items())[:100])

# Create hash table and insert
ht = FixedHashTable(200)  # Size 200 for 100 words (low load factor)
for word, freq in first_100.items():
    ht.insert(word, freq)

# Test operations
print(f"Inserted {len(first_100)} words")
print(f"Get 'the': {ht.get('the')}")
print(f"First: {ht.get_first()}")
print(f"Last: {ht.get_last()}")
```
