## Demo

```python
from binance_trades_client import BinanceTradesClient
```

```python
client = BinanceTradesClient()

# Stream and print first 3 trades
trade_count = [0]

def handle_trade(trade):
    trade_count[0] += 1
    if trade_count[0] <= 3:
        client.print_trade(trade)

print("=== Streaming Trades (5 seconds) ===")
client.stream_trades(callback=handle_trade, duration=2)
print(f"\nReceived {trade_count[0]} trades")
```
